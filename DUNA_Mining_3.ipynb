```python
# @title Step 2: Mint property
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

import uuid
import time
import requests
from IPython.display import display, HTML, IFrame, clear_output
import gspread
import datetime
import json

# ========== Configuration ==========
SPREADSHEET_ID = "1Lie8a1j12SxSTFnBGtIC3gFeOCCbSgHiEWxhaFzFqzM"
WEBHOOK_POLL_URL = "https://webhook.site/token/237852a9-da49-4845-8cb3-323331fe40d5/requests?sorting=newest&per_page=100"

# Hardcode the constants you want to pass to the app
SEED_DATA_GROUP_CID = "bafkreihozyn5qubfuw525j5f67bnnnph75xj5wue2mhshwhcxbthkefzti"
COUNTY_DATA_GROUP_CID = "bafkreicl4paz35lrgav4cniflcj6nbphp3fwdapv2lcbe2aogjrcqpyjpq"
HTML_LINK_OVERRIDE = ""  # Optional fallback HTML link
PROCESSING_TIMEOUT_MIN = 30

current_parcel: dict[str, str] | None = None


def info_box(message: str, kind: str = "info"):
    colors = {
        "info": ("#e3f2fd", "#1976d2"),
        "error": ("#ffebee", "#c62828"),
    }
    bg, border = colors.get(kind, colors["info"])
    display(HTML(
        f"""
        <div style="padding: 12px; border: 2px solid {border}; background: {bg};
                    border-radius: 6px; margin: 10px 0;">
            {message}
        </div>
        """
    ))


def open_sheet():
    from google.colab import auth
    from google.auth import default

    auth.authenticate_user()
    creds, _ = default(scopes=["https://www.googleapis.com/auth/spreadsheets"])
    gc = gspread.authorize(creds)
    ws = gc.open_by_key(SPREADSHEET_ID).sheet1
    headers = ws.row_values(1)
    return ws, headers


def col_idx(headers: list[str], name: str) -> int | None:
    return headers.index(name) if name in headers else None


def release_processing(ws, headers, row_number: int):
    status_col = col_idx(headers, "Status")
    ts_col = col_idx(headers, "Status TimeStamp")
    if row_number and status_col is not None and ts_col is not None:
        ws.update_cell(row_number, status_col + 1, "available")
        ws.update_cell(row_number, ts_col + 1, "")


def get_next_parcel_from_sheet():
    global current_parcel
    ws, headers = open_sheet()

    parcel_col = col_idx(headers, "parcel_id")
    seed_col = col_idx(headers, "seed_hash")
    county_col = col_idx(headers, "county_hash")
    status_col = col_idx(headers, "Status")
    ts_col = col_idx(headers, "Status TimeStamp")
    txn_col = col_idx(headers, "Transaction ID")

    if None in (parcel_col, seed_col, county_col, status_col, ts_col, txn_col):
        info_box("Sheet must include parcel_id, seed_hash, county_hash, Status, Status TimeStamp, Transaction ID columns.", "error")
        return None, ws, headers

    all_rows = ws.get_all_values()
    now = datetime.datetime.now(datetime.timezone.utc)
    now_iso = now.isoformat()

    for idx, row in enumerate(all_rows[1:], start=2):
        parcel = row[parcel_col] if len(row) > parcel_col else ""
        seed = row[seed_col] if len(row) > seed_col else ""
        county = row[county_col] if len(row) > county_col else ""
        status = row[status_col] if len(row) > status_col else ""
        txn = row[txn_col] if len(row) > txn_col else ""
        started = row[ts_col] if len(row) > ts_col else ""

        if not parcel or not seed or not county or txn:
            continue

        if status == "processing" and started:
            try:
                started_time = datetime.datetime.fromisoformat(started.replace("Z", "+00:00"))
                elapsed = (now - started_time).total_seconds() / 60
                if elapsed <= PROCESSING_TIMEOUT_MIN:
                    continue
            except Exception:
                pass
            ws.update_cell(idx, status_col + 1, "available")
            ws.update_cell(idx, ts_col + 1, "")
            status = "available"

        if status and status not in ("available", "processing"):
            continue

        ws.update_cell(idx, status_col + 1, "processing")
        ws.update_cell(idx, ts_col + 1, now_iso)
        current_parcel = {
            "row_number": idx,
            "parcel_id": parcel,
            "property_cid": seed,
            "data_cid": county,
        }
        return current_parcel, ws, headers

    info_box("No available parcels found in the sheet.", "error")
    return None, ws, headers


# ========== Main Function ==========
def mint_and_log():
    verify_acknowledgment()

    parcel, ws, headers = get_next_parcel_from_sheet()
    if not parcel:
        return

    property_cid = parcel["property_cid"]
    data_cid = parcel["data_cid"]
    parcel_id = parcel["parcel_id"]

    if not property_cid or not data_cid:
        info_box("Missing seed_hash or county_hash for the selected row.", "error")
        release_processing(ws, headers, parcel.get("row_number"))
        return

    seed_data_group_cid = SEED_DATA_GROUP_CID
    county_data_group_cid = COUNTY_DATA_GROUP_CID
    html_link = HTML_LINK_OVERRIDE

    if not seed_data_group_cid or not county_data_group_cid:
        info_box("Please set SEED_DATA_GROUP_CID and COUNTY_DATA_GROUP_CID constants.", "error")
        release_processing(ws, headers, parcel.get("row_number"))
        return

    session_id = str(uuid.uuid4())[:8]
    unique_id = f"{property_cid}_{session_id}"
    data_param = f"{property_cid},{seed_data_group_cid},{county_data_group_cid},{data_cid}"
    minting_url = f"https://oracle.elephant.xyz/?session-id={unique_id}&data={data_param}"

    display(HTML(f"""
        <div style="padding: 15px; background-color: #e3f2fd; border: 2px solid #1976d2;
                    border-radius: 8px; margin: 15px 0;">
            <h3>üîó Mint Your Property</h3>
            <p>Complete the minting process below. If the app doesn't load,
               <a href="{minting_url}" target="_blank">click here to open in a new tab</a>.</p>
        </div>
    """))

    display(IFrame(src=minting_url, width="100%", height="600"))

    display(HTML("""
        <div style="padding: 15px; background-color: #fff3e0; border: 2px solid #ff9800;
                    border-radius: 8px; margin: 15px 0;">
            <h3>‚è≥ Waiting for minting confirmation (Please don't close the window)...</h3>
            <p>Complete the minting process above. This cell will automatically detect when it's done.</p>
        </div>
    """))

    wallet_address = ""
    transaction_hash = ""
    max_wait = 600
    poll_interval = 5
    waited = 0
    WEBHOOK_API_KEY = "2dd2ca9b-5914-4ed9-83a9-c34365e6a0dc"

    while waited < max_wait:
        try:
            api_headers = {"Api-Key": WEBHOOK_API_KEY}
            response = requests.get(WEBHOOK_POLL_URL, headers=api_headers, timeout=10)
            if response.status_code == 200:
                data = response.json()
                for req in data.get("data", []):
                    content = req.get("content", "")
                    if isinstance(content, str):
                        try:
                            content = json.loads(content)
                        except Exception:
                            continue

                    if content.get("sessionId") == unique_id:
                        print(content)
                        wallet_address = content.get("userAddress", "")
                        transaction_hash = content.get("transactionHash", "")
                        break

                if wallet_address and transaction_hash:
                    break
        except Exception:
            pass

        time.sleep(poll_interval)
        waited += poll_interval
        print(f"\r‚è≥ Waiting... {waited}s", end="", flush=True)

    clear_output(wait=True)

    if not wallet_address or not transaction_hash:
        info_box("Timeout: No minting confirmation received", "error")
        release_processing(ws, headers, parcel.get("row_number"))
        return

    display(HTML(f"""
        <div style="padding: 15px; background-color: #e8f5e9; border: 2px solid #388e3c;
                    border-radius: 8px; margin: 15px 0;">
            <h3>‚úÖ Minting Confirmed!</h3>
            <p><b>Wallet Address:</b> {wallet_address}</p>
            <p><b>Transaction Hash:</b> {transaction_hash}</p>
            <p><a href="https://polygonscan.com/tx/{transaction_hash}" target="_blank">View on PolygonScan ‚Üí</a></p>
        </div>
    """))

    def update_if(col_name: str, value: str):
        col = col_idx(headers, col_name)
        if col is not None:
            ws.update_cell(parcel["row_number"], col + 1, value)

    timestamp_now = datetime.datetime.now(datetime.timezone.utc).isoformat()

    update_if("ID", unique_id)
    update_if("Status", "completed")
    update_if("Status TimeStamp", timestamp_now)
    update_if("Acknowledgment TimeStamp", acknowledgment_data.get("timestamp", ""))
    update_if("Name", acknowledgment_data.get("full_name", ""))
    update_if("Email", acknowledgment_data.get("email", ""))
    update_if("Completetion TimeStamp", timestamp_now)
    update_if("Wallet Address", wallet_address)
    update_if("Transaction ID", transaction_hash)
    update_if("HTML link", html_link)


# Run
mint_and_log()
```
